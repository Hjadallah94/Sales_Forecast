#### Changing the Format for the "Transaction Date" field, this was done to produce the model

#### Merging the "Product" and "Sales_Daily" data sets together

```python
result = pd.merge(df3, df4, on='IteE ID', how='inner')

print(result)
```

#### Copying the date frame to csv file, to use them in different notebooks

```python
result2 = result[["Category Code", "Category Name", "Transaction Date", "Net Gross Profit"]]
```

```python
result2
```

```python
"""
The next step is performed to sort the data from the first transaction date to the last

"""

result2.sort_values("Transaction Date", inplace=True)
result2.reset_index(drop=True, inplace=True)

result2
```

```python
Womens_result2 = result2[result2['Category Code'] == 100110]
Womens_result2
```

```python
Mens_result2 = result2[result2['Category Code'] == 100120]
Mens_result2
```

```python
Accessories_result2 = result2[result2['Category Code'] == 100100]
Accessories_result2
```

```python
Womens_result2_Grouped = Womens_result2.groupby("Transaction Date")
Womens_result2_Grouped
```

```python
Mens_result2_Grouped = Mens_result2.groupby("Transaction Date")
Mens_result2_Grouped
```

```python
Accessories_result2_Grouped = Accessories_result2.groupby("Transaction Date")
Accessories_result2_Grouped
```

```python
Womens_result2_Grouped_df = Womens_result2_Grouped["Net Gross Profit"].sum()
Womens_result2_Grouped_df = pd.DataFrame({"Transaction Date": Womens_result2_Grouped_df.index, "Net Gross Profit": Womens_result2_Grouped_df.values})
```

```python
Womens_result2_Grouped_df
```

```python
Mens_result2_Grouped_df = Mens_result2_Grouped["Net Gross Profit"].sum()
Mens_result2_Grouped_df = pd.DataFrame({"Transaction Date": Mens_result2_Grouped_df.index, "Net Gross Profit": Mens_result2_Grouped_df.values})
```

```python
Mens_result2_Grouped_df
```

```python
Accessories_result2_Grouped_df = Accessories_result2_Grouped["Net Gross Profit"].sum()
Accessories_result2_Grouped_df = pd.DataFrame({"Transaction Date": Accessories_result2_Grouped_df.index, "Net Gross Profit": Accessories_result2_Grouped_df.values})
```

```python
Accessories_result2_Grouped_df
```

```python
Womens_result2_Grouped_df.to_csv('Womens_NetGrossProfit.csv', index=False)
Mens_result2_Grouped_df.to_csv('Mens_NetGrossProfit.csv', index=False)
Accessories_result2_Grouped_df.to_csv('Accessories_NetGrossProfit.csv', index=False)
```
